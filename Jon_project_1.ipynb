```python
# Set style
sns.set(style="whitegrid")

# Load datasets
covid_data = pd.read_csv('covid_data.csv')
happiness_data = pd.read_csv('WorldHappinessIndex2013-2023.csv')

# Display the first rows
print("COVID-19 Data Columns:")
print(covid_data.columns)
print("\nWorld Happiness Data Columns:")
print(happiness_data.columns)

# Filter for specific countries
countries = ['United States', 'Afghanistan', 'Finland']
covid_data_filtered = covid_data[covid_data['location'].isin(countries)].copy()
happiness_data_filtered = happiness_data[happiness_data['Country'].isin(countries)].copy()

# Convert the date column to datetime
covid_data_filtered['date'] = pd.to_datetime(covid_data_filtered['date'])

# Define a color palette
colors = sns.color_palette("Set2", len(countries))

# Plotting COVID-19 total cases over time for each country with a logarithmic scale
plt.figure(figsize=(14, 7))
for i, country in enumerate(countries):
    country_data = covid_data_filtered[covid_data_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=f'Total COVID-19 Cases in {country}', color=colors[i], linewidth=1)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Cases', fontsize=14)
plt.title('Total COVID shown over time Cases', fontsize=18, fontweight='bold')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.yscale('log')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12, rotation=45)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
plt.show()

# Make sure column names for happiness score
if 'Happiness Score' in happiness_data_filtered.columns:
    happiness_score_col = 'Happiness Score'
elif 'Happiness.Score' in happiness_data_filtered.columns:
    happiness_score_col = 'Happiness.Score'
else:
    print("Happiness score column not found!")
    happiness_score_col = None

# Plotting 
if happiness_score_col:
    plt.figure(figsize=(14, 7))
    for i, country in enumerate(countries):
        country_data = happiness_data_filtered[happiness_data_filtered['Country'] == country]
        plt.plot(country_data['Year'], country_data[happiness_score_col], label=f'Happiness Score in {country}', color=colors[i], linewidth=1)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Happiness Score', fontsize=14)
    plt.title('Happiness Score Over the Years', fontsize=18, fontweight='bold')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Load the happiness dataset
happiness_data = pd.read_csv('WorldHappinessIndex2013-2023.csv')

# Filter for specific countries
countries = ['United States', 'Finland', 'Afghanistan']
happiness_data_filtered = happiness_data[happiness_data['Country'].isin(countries)].copy()

# Do correct column names
if 'Index' in happiness_data_filtered.columns:
    index_col = 'Index'
else:
    print("Index column not found!")
    index_col = None

# Plotting Index
if index_col:
    plt.figure(figsize=(14, 7))
    palette = sns.color_palette("Set3", len(countries))
    
    for i, country in enumerate(countries):
        country_data = happiness_data_filtered[happiness_data_filtered['Country'] == country]
        plt.plot(country_data['Year'], country_data[index_col], marker='o', label=f'Index in {country}', color=palette[i])
        plt.fill_between(country_data['Year'], country_data[index_col], alpha=0.1, color=palette[i])
    
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Index', fontsize=14)
    plt.title('Index Over the Years', fontsize=18, fontweight='bold')
    plt.legend(title='Country', title_fontsize='13', fontsize='11')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    # Adding a fancy background
    sns.despine()
    plt.gca().set_facecolor('#f0f0f0')
    
    plt.show()
```
