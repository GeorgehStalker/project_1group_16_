```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the COVID-19 dataset
covid_data_path = 'C:/Users/georg/git repositories/project 1 group 16/project_1group_16_/owid-covid-data (1).csv'
covid_data = pd.read_csv(covid_data_path)

# Convert 'date' to datetime format for better handling
covid_data['date'] = pd.to_datetime(covid_data['date'])

# Extract the year from the date
covid_data['year'] = covid_data['date'].dt.year

# Group by year and sum up the total cases globally
global_cases_data = covid_data.groupby('year').agg({'total_cases': 'sum'}).reset_index()

# Define significant events and their corresponding years for annotation
events = {
    2020: 'Pandemic Outbreak',
    2020: 'First Lockdowns Worldwide',
    2021: 'Vaccine Rollout Begins',
    2021: 'Delta Variant Emergence',
    2022: 'Omicron Variant Emergence',
    2023: 'Endemic Transition Discussions'
}

# Create a plot for Total Cases over years worldwide with a COVID-19 theme
plt.figure(figsize=(14, 8))
sns.set_theme(style="darkgrid")

# Plotting Total Cases
sns.lineplot(data=global_cases_data, x='year', y='total_cases', color='red', marker='o', linewidth=2.5)

# Adding titles and labels
plt.title('Global Trends in Total COVID-19 Cases Over the Years', fontsize=18, fontweight='bold', color='darkred')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Cumulative Total Cases', fontsize=16)
plt.xticks(global_cases_data['year'], rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Adding annotations for significant events
for year, event in events.items():
    # Find the corresponding value for the year to annotate
    cases_value = global_cases_data[global_cases_data['year'] == year]['total_cases'].values[0]
    plt.annotate(event, (year, cases_value), textcoords="offset points", xytext=(0, 10), ha='center',
                 fontsize=10, color='black', backgroundcolor='yellow')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the World Happiness Index dataset
happiness_data_path = 'C:/Users/georg/git repositories/project 1 group 16/project_1group_16_/WorldHappinessIndex2013-2023.csv'
happiness_data = pd.read_csv(happiness_data_path)

# Display the first few rows of the dataset to understand its structure
print(happiness_data.head())

# Group by year and calculate the average happiness score
average_happiness = happiness_data.groupby('Year').agg({'Index': 'mean'}).reset_index()

# Define significant events and their corresponding years for annotation
events = {
    2015: 'First Data Published',
    2016: 'Increasing Awareness',
    2017: 'Global Collaboration Initiatives',
    2018: 'Mental Health Awareness',
    2019: 'Growing Research Focus',
    2020: 'COVID-19 Pandemic Begins',
    2021: 'Pandemic Response Initiatives',
    2022: 'End of COVID-19 Emergency in Many Countries',
    2023: 'Post-Pandemic Adjustments'
}

# Create a plot for Happiness Index scores over the years
plt.figure(figsize=(14, 8))
sns.set_theme(style="darkgrid")

# Plotting the Happiness Index
sns.lineplot(data=average_happiness, x='Year', y='Index', marker='o', color='blue', linewidth=2.5)

# Adding titles and labels
plt.title('Global Happiness Index Over the Years', fontsize=18, fontweight='bold', color='green')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Average Happiness Score', fontsize=16)
plt.xticks(average_happiness['Year'], rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Adding annotations for significant events
for year, event in events.items():
    # Find the corresponding value for the year to annotate
    index_value = average_happiness[average_happiness['Year'] == year]['Index'].values[0]
    plt.annotate(event, (year, index_value), textcoords="offset points", xytext=(0, 10), ha='center',
                 fontsize=10, color='yellow', backgroundcolor='green')

plt.grid(True)
plt.tight_layout()
plt.show()
```
